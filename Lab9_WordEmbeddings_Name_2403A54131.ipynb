```python
import gensim
from gensim.models import KeyedVectors
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import gensim.downloader as api
model = api.load("word2vec-google-news-300")
print("Vocabulary size:", len(model.key_to_index))
word ="dog"
vector = model[word]
print(f"Vector length", len(vector))
print("1st 10 values of the vector:", vector[:10])
```

```python
model2 = api.load("glove-wiki-gigaword-100")
print("Vocabulary size:", len(model2.key_to_index))
vector2 = model2[word]
print(f"Vector length", len(vector2))
print("1st 10 values of the vector:", vector2[:10])
```

```python
word_pairs = [("king", "queen"),("man", "woman"),("boy", "girl"),("cat", "dog"),("car", "bicycle")]
print("word similarity scores")
for w1, w2 in word_pairs:
    sim = model.similarity(w1, w2)
    print(f"{w1} - {w2}: {sim:.4f}")
```

```python
chosen_words=["woman", "boy", "girl", "cat", "dog"]
for w in chosen_words:
    print(f"top similar words to {w}:\n")
    similar_words = model2.most_similar(w, topn=5)
    for word, similarity in similar_words:
        print(f"  {word}: {similarity:.4f}")
```

```python
result1 = model.most_similar(positive=["king", "woman"], negative=["man"], topn=5)
result2 = model.most_similar(positive=["prince", "woman"], negative=["man"], topn=5)
result= model.most_similar(positive=["doctor", "hospital"], negative=["patient"], topn=5)
print("Results for 'king - man + woman':", result1)
print("Results for 'prince - man + woman':", result2)
print("Results for 'doctor - patient + hospital':", result)
```
