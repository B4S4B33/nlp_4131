```python
import pandas as pd 
data=pd.read_csv('sentences_ngram.csv')
data.head()
```

```python
import collections
combined_sentences = ' '.join(data['sentence'])
words = combined_sentences.split()
word_counts = collections.Counter(words)
print("unigram_counts: ")
for word, count in word_counts.items():
    print(f"{word}: {count}")
V=len(word_counts)
print("Vocabulary size (V):", V)
```

```python
bigram=[]
for i in range(len(words)-1):
    bigram.append((words[i], words[i+1]))
bigram_counts = collections.Counter(bigram)
print("\nbigram_counts: ")
for bigram, count in bigram_counts.items():
    print(f"{bigram}: {count}")
```

```python
trigram=[]
for i in range(len(words)-2):
    trigram.append((words[i], words[i+1], words[i+2]))
trigram_counts = collections.Counter(trigram)
print("\ntrigram_counts: ")
for trigram, count in trigram_counts.items():
    print(f"{trigram}: {count}")
```

```python
def predict_bigram(word_sequence,bigram_counts,unigram_counts):
    word_in_seq = word_sequence.lower().split()
    if not word_in_seq:
        return "please enter a word sequence"
    last_word = word_in_seq[-1]
    Candidate_words = {}
    for (w1, w2), count in bigram_counts.items():
        if w1 == last_word:
            Candidate_words[w2] = count
    if not Candidate_words:
        return "No predictions available for the given word sequence."
    last_word_count = unigram_counts.get(last_word, 0)
    if last_word_count == 0:
        return "No predictions available for the given word sequence."
    predicted_words = None
    max_prob = -1
    for next_word, bigram_count in Candidate_words.items():
        prob = bigram_count /last_word_count
        if prob > max_prob:
            max_prob = prob
            predicted_words = next_word
    return predicted_words
seq= input("Enter a word sequence: ")
predicted_word = predict_bigram(seq, bigram_counts, word_counts)
print(f"Predicted next word: {predicted_word}")
```

```python
def predict_trigram(word_sequence,trigram_counts,bigram_counts):
    word_in_seq = word_sequence.lower().split()
    if len(word_in_seq) < 2:
        return "Please enter at least two words for trigram prediction."
    last_two_words = tuple(word_in_seq[-2:])
    Candidate_words = {}
    for (w1, w2, w3), count in trigram_counts.items():
        if (w1, w2) == last_two_words:
            Candidate_words[w3] = count
    if not Candidate_words:
        return "No predictions available for the given word sequence."
    bigram_count = bigram_counts.get(last_two_words, 0)
    if bigram_count == 0:
        return "No predictions available for the given word sequence."
    predicted_word = None
    max_prob = -1
    for next_word, trigram_count in Candidate_words.items():
        prob = trigram_count / bigram_count
        if prob > max_prob:
            max_prob = prob
            predicted_word = next_word
    return predicted_word
seq2= input("Enter a word sequence: ")
predicted_word2 = predict_trigram(seq2, trigram_counts, bigram_counts)
print(f"Predicted next word: {predicted_word2}")
```

```python
def predict_bigram_laplace(word_sequence,bigram_counts,unigram_counts,V):
    word_in_seq = word_sequence.lower().split()
    if not word_in_seq:
        return "please enter a word sequence"
    last_word = word_in_seq[-1]
    Candidate_words = {}
    for (w1, w2), count in bigram_counts.items():
        if w1 == last_word:
            Candidate_words[w2] = count
    if not Candidate_words:
        return "No predictions available for the given word sequence."
    last_word_count = unigram_counts.get(last_word, 0)
    predicted_word = None
    max_prob = -1
    for next_word in unigram_counts.keys():
        bigram_count = Candidate_words.get(next_word, 0)
        prob = (bigram_count + 1) / (last_word_count + V)
        if prob > max_prob:
            max_prob = prob
            predicted_word = next_word
    return predicted_word
seq3= input("Enter a word sequence: ")
predicted_word3 = predict_bigram_laplace(seq3, bigram_counts, word_counts,V)
print(f"Predicted next word with Laplace smoothing: {predicted_word3}")
```
